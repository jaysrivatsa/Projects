```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Superstore_Sales.xls')
```

```python
df = df[['Customer ID','Sub-Category']]
```

```python
df
```

```python
group1 = df.groupby(['Customer ID','Sub-Category']).size().unstack()
group1
```

```python
group1 = group1.fillna(0)
```

```python
group1.columns
```

```python
group2 = group1.values.astype('bool').astype('int')
group2 = pd.DataFrame(group2, columns = group1.columns, index = group1.index)
```

```python
group2
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_itemsets = apriori(group2, min_support=0.2, use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets.sort_values(['support'],ascending=[0])
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
rules[(rules.confidence >= 0.8)]
```
